# Replacing the missing values in numerical columns with median and categorical column with mode 

```python
for i in ["Title I Served School" , "Mathematics Growth Score (GS)" , "Mathematics Growth Status" , "Mathematics Growth Index"]:
    if df[i].dtype == 'O':
        df[i].fillna(df[i].mode()[0] , inplace=True)
    else:
        df[i].fillna(df[i].median() , inplace=True)
```
